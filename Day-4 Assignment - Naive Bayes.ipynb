```python
#Encoding Labels

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Disease'] = label_encoder.fit_transform(df['Disease'])
df['Fever'] = label_encoder.fit_transform(df['Fever'])
df['Cough'] = label_encoder.fit_transform(df['Cough'])
df['Fatigue'] = label_encoder.fit_transform(df['Fatigue'])
df['Difficulty Breathing'] = label_encoder.fit_transform(df['Difficulty Breathing'])
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Blood Pressure'] = label_encoder.fit_transform(df['Blood Pressure'])
df['Cholesterol Level'] = label_encoder.fit_transform(df['Cholesterol Level'])
df['Outcome Variable'] = label_encoder.fit_transform(df['Outcome Variable'])
```

# Assign X & Y

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:].values
```

```python
x.head(2)
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

# Spliting data into training and testing

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .2,random_state = 2)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

# Gaussian NB Model

```python
from sklearn.naive_bayes import GaussianNB

gauss_nb = GaussianNB()
gauss_nb.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred = gauss_nb.predict(xtest)
print('Predicted value is: \t',ypred)
print()
print('Testing completed')
print('='*75)
```

# Performance Measure

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print('*'*15 + 'Performance Measures on Gaussian Navie Bayes' + '*'*16)
print('Accuracy score:\t\n',accuracy_score(ytest,ypred))
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred))
print()
print('='*75)
```

```python
sns.heatmap(confusion_matrix(ytest,ypred),annot = True)
```

```python
from sklearn.naive_bayes import MultinomialNB

mul_nb = MultinomialNB()
mul_nb.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred = mul_nb.predict(xtest)
print('Predicted value is: \t',ypred)
print()
print('Testing completed')
print('='*75)

print('*'*14 + 'Performance Measures on Multinomial Navie Bayes' + '*'*14)
print('Accuracy score:\t\n',accuracy_score(ytest,ypred))
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred))
print()
print('='*75)
```
