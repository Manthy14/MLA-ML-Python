# Assign X & Y

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:].values

print('X shape:\t',x.shape)
print()
print('Y shape:\t',y.shape)
```

# Split the data between training and testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2,random_state=10)
```

# Build NB - Gaussian Model

```python
from sklearn.naive_bayes import GaussianNB

gauss_nb = GaussianNB()
gauss_nb.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred_gauss= gauss_nb.predict(xtest)
print('Predicted value is: \t',ypred_gauss)
print()
print('Testing completed')
print('='*75)

# Performance Measure

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print('*'*15 + 'Performance Measures on Gaussian Navie Bayes' + '*'*16)
accuracy_score_gauss = accuracy_score(ytest,ypred_gauss)
print('Accuracy score:\t\n',accuracy_score_gauss)
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred_gauss))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred_gauss))
print()
print('='*75)
```

# Build NB - Multinomial Model

```python
from sklearn.naive_bayes import MultinomialNB

mul_nb = MultinomialNB()
mul_nb.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred_mul = mul_nb.predict(xtest)
print('Predicted value is: \t',ypred_mul)
print()
print('Testing completed')
print('='*75)

#Performance Measure

print('*'*14 + 'Performance Measures on Multinomial Navie Bayes' + '*'*14)
accuracy_score_mul = accuracy_score(ytest,ypred_mul)
print('Accuracy score:\t\n',accuracy_score_mul)
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred_mul))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred_mul))
print()
print('='*75)
```

# Build NB - Bernoulli Model

```python
from sklearn.naive_bayes import BernoulliNB

ber_nb = BernoulliNB()
ber_nb.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred_ber = ber_nb.predict(xtest)
print('Predicted value is: \t',ypred_ber)
print()
print('Testing completed')
print('='*75)

#Performace Measure

print('*'*15 + 'Performance Measures on Bernoulli Navie Bayes' + '*'*15)
accuracy_score_ber = accuracy_score(ytest,ypred_ber)
print('Accuracy score:\t\n',accuracy_score_ber)
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred_ber))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred_ber))
print()
print('='*75)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred_knn = knn.predict(xtest)
print('Predicted value is: \t',ypred_knn)
print()
print('Testing completed')
print('='*75)

#Performace Measure

print('*'*15 + 'Performance Measures on KNN' + '*'*15)
accuracy_score_knn = accuracy_score(ytest,ypred_knn)
print('Accuracy score:\t\n',accuracy_score_knn)
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred_knn))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred_knn))
print()
print('='*75)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()

dt.fit(xtrain,ytrain)
print('Training Completed')
print()
ypred_dt = dt.predict(xtest)
print('Predicted value is: \t',ypred_dt)
print()
print('Testing completed')
print('='*75)

#Performace Measure

print('*'*15 + 'Performance Measures on Decision Tree' + '*'*15)
accuracy_score_dt = accuracy_score(ytest,ypred_dt)
print('Accuracy score:\t\n',accuracy_score_dt)
print()
print('='*75)
print()
print('Confusion matrix:\t\n',confusion_matrix(ytest,ypred_dt))
print()
print('='*75)
print()
print('Classification report is:\t\n',classification_report(ytest,ypred_dt))
print()
print('='*75)
```

# Applying Ensembling Model

# 1. BaggingClassifier

```python
#Bagging Classifier

from sklearn.ensemble import BaggingClassifier
bagc = BaggingClassifier(base_estimator=gauss_nb,n_estimators=5,random_state=1) #Instantiate Bagging Classifier

# training the model
bagc.fit(xtrain,ytrain)

#testing the model
ypred_bagc = bagc.predict(xtest)

print('*'*75)
print('Confusion Matrix:\t\n',confusion_matrix(ytest,ypred_bagc))
print('-'*75)
accuracy_bagc = accuracy_score(ytest,ypred_bagc)
print('Accuracy score:\t\n',accuracy_bagc)
print('-'*75)
print('Classification report\t\n:',classification_report(ytest,ypred_bagc))
print('*'*75)
```

# 2. RandomForest Classifier Model

```python
#RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier(n_estimators=10,random_state=1)
# training the model
rand.fit(xtrain,ytrain)

#testing the model
ypred_rand = rand.predict(xtest)

print('*'*75)
print('Confusion Matrix:\t\n',confusion_matrix(ytest,ypred_rand))
print('-'*75)
accuracy_rand = accuracy_score(ytest,ypred_rand)
print('Accuracy score:\t\n',accuracy_rand)
print('-'*75)
print('Classification report\t\n:',classification_report(ytest,ypred_rand))
print('*'*75)

```

```python
#RandomForest Classifier with puring

from sklearn.ensemble import RandomForestClassifier
rand_1 = RandomForestClassifier(n_estimators=10,random_state=1,max_features=5)
# training the model
rand_1.fit(xtrain,ytrain)

#testing the model
ypred_rand_1 = rand_1.predict(xtest)

print('*'*75)
print('Confusion Matrix:\t\n',confusion_matrix(ytest,ypred_rand_1))
print('-'*75)
accuracy_rand_1 = accuracy_score(ytest,ypred_rand_1)
print('Accuracy score:\t\n',accuracy_rand_1)
print('-'*75)
print('Classification report\t\n:',classification_report(ytest,ypred_rand_1))
print('*'*75)

```

# 3. ExtraTrees Model

```python
from sklearn.ensemble import ExtraTreesClassifier
extra_tree = ExtraTreesClassifier(n_estimators=10,bootstrap = True,verbose=2)
# training the model
extra_tree.fit(xtrain,ytrain)

#testing the model
ypred_extra = extra_tree.predict(xtest)

print('*'*75)
print('Confusion Matrix:\t\n',confusion_matrix(ytest,ypred_extra))
print('-'*75)
accuracy_extra = accuracy_score(ytest,ypred_extra)
print('Accuracy score:\t\n',accuracy_extra)
print('-'*75)
print('Classification report\t\n:',classification_report(ytest,ypred_extra))
print('*'*75)
```
